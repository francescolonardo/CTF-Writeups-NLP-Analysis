## Model testing (Entity Extraction)

Testing model's predictions on an unseen sample writeup text.
We need to load the model and run it on our prepared text.

```python
import spacy
nlp = spacy.load("drive/MyDrive/model/model-best")

file_path = "drive/MyDrive/sample_writeup.txt"
with open(file_path, 'r') as f:
  text = f.read()

for doc in nlp.pipe(text, disable=["tagger"]):
  print([(ent.start, ent.text, ent.label_) for ent in doc.ents])
```
